### 2. What is the standard deviation of Sp. Def. in the dataset ?

```python
pk['Sp. Def'].std()
```

### 3. What percentage (upto 3 decimal places) of pokemons are legendary ? 

```python
pk['Legendary'].value_counts(normalize=True) * 100
```

### 4. Name the pokemon(s) with Maximum Defense. 

```python
print(pk[pk.Defense == pk.Defense.max()])
```

### 5. Which poison pokemon has the strongest attack ?

```python
pk.loc[pk['Type 2'] == 'Poison', 'Attack'].sort_values(ascending = False).head(1)
```

### 6. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
pk[(pk['Legendary'] == False) & (pk['Generation'] == 4) & (pk['Type 1'] == 'Ghost')].sort_values
```

### 7. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
pk[(pk['Legendary'] == False) & (pk['Defense']>78) & (pk['Attack']<84)].count()
```
